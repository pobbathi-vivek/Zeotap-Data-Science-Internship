```python



# Define price scales (multiples of 50)
price_scale = np.arange(0, products_df['Price'].max() + 50, 50)

# Function to assign products to the price scale
def assign_price_scale(price):
    return pd.cut([price], bins=price_scale, right=False, labels=[f"{int(price_scale[i])}-{int(price_scale[i+1]-1)}" for i in range(len(price_scale)-1)])[0]

# Add a new column 'PriceScale' to categorize products into price ranges
products_df['PriceScale'] = products_df['Price'].apply(assign_price_scale)

# Now we will group by 'Category' and 'PriceScale' and get the count of products and their names
product_count_by_scale = products_df.groupby(['Category', 'PriceScale']).agg(
    ProductCount=('ProductName', 'count'),
    ProductNames=('ProductName', lambda x: ', '.join(x))
).reset_index()

# Convert the 'PriceScale' to an ordered category to sort correctly
price_scale_order = pd.Categorical(product_count_by_scale['PriceScale'], categories=product_count_by_scale['PriceScale'].unique(), ordered=True)
product_count_by_scale['PriceScale'] = price_scale_order

# Sort the DataFrame by 'PriceScale' (and optionally by 'Category' if needed)
product_count_by_scale = product_count_by_scale.sort_values(by=['PriceScale', 'Category'], ascending=[True, True])

# Display the result
product_count_by_scale

```

```python


# Set up the figure
plt.figure(figsize=(12, 8))

# Create a bar plot for product count by price scale and category
ax = sns.barplot(data=product_count_by_scale, x='PriceScale', y='ProductCount', hue='Category', palette='viridis')

# Add numbers on top of the bars
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',  # Display the height of the bar
        (p.get_x() + p.get_width() / 2, p.get_height()),  # Position at the top center of each bar
        ha='center', va='bottom', fontsize=10, color='black'  # Formatting the text
    )

# Add titles and labels
plt.title('Product Count by Price Scale and Category')
plt.xlabel('Price Scale')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.show()
```

```python
# Pivot the data to get 'Category' as columns, 'PriceScale' as index, and 'ProductCount' as values
pivoted_data = product_count_by_scale.pivot_table(
    index='PriceScale', columns='Category', values='ProductCount', aggfunc='sum', fill_value=0
)

# Set up the figure
ax = pivoted_data.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 8))

# Add numbers on top of the bars in the stacked bar plot
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(
            f'{int(height)}',  # Display the height of the bar
            (p.get_x() + p.get_width() / 2, p.get_y() + height / 2),  # Position at the center of each stacked segment
            ha='center', va='center', fontsize=10, color='black'  # Formatting the text
        )

# Add titles and labels
plt.title('Stacked Product Count by Price Scale and Category')
plt.xlabel('Price Scale')
plt.ylabel('Number of Products')

# Show the plot
plt.show()
```

```python
# Extract year from TransactionDate
merged_df['TransactionYear'] = merged_df['TransactionDate'].dt.year

# Filter data for 2023 and 2024
sales_2023 = merged_df[merged_df['TransactionYear'] == 2023]
sales_2024 = merged_df[merged_df['TransactionYear'] == 2024]
```

```python
# Function to plot monthly sales by continent
def plot_monthly_sales_by_continent(data, year):
    data['TransactionMonth'] = data['TransactionDate'].dt.month
    monthly_sales = data.groupby(['TransactionMonth', 'Region'])['TotalValue'].sum().unstack()
    
    plt.figure(figsize=(14, 8))
    monthly_sales.plot(kind='bar', stacked=True, colormap='viridis')
    plt.title(f'Monthly Sales by Continent for {year}')
    plt.xlabel('Month')
    plt.ylabel('Total Sales (USD)')
    plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
    plt.legend(title='Continent')
    plt.show()

# Plot for 2023
plot_monthly_sales_by_continent(sales_2023, 2023)

# Plot for 2024
plot_monthly_sales_by_continent(sales_2024, 2024)
```

```python
# Function to plot monthly sales for all continents combined
def plot_monthly_sales_all_continents(data, year):
    data['TransactionMonth'] = data['TransactionDate'].dt.month
    monthly_sales = data.groupby('TransactionMonth')['TotalValue'].sum()
    
    plt.figure(figsize=(14, 8))
    sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
    plt.title(f'Monthly Sales for All Continents Combined for {year}')
    plt.xlabel('Month')
    plt.ylabel('Total Sales (USD)')
    plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
    plt.show()

# Plot for 2023
plot_monthly_sales_all_continents(sales_2023, 2023)

# Plot for 2024
plot_monthly_sales_all_continents(sales_2024, 2024)
```
